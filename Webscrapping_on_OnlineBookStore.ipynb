```python
import pandas as pd
import requests
import bs4
```

What we need:
1. Category_Name
2. urls(images)
3. title
4. Tab_Index

```python

#creating empty lists
category_names = []
urls_images = []
title = []
tab_index = []
```

```python
sites = 'https://openlibrary.org/'
```

```python
print(sites)
```

```python
page=requests.get(sites)
soup=bs4.BeautifulSoup(page.text,'html.parser')
```

```python
print(soup)
```

```python
for t in soup.findAll('title'):
  print(t)
```

```python
print(soup.prettify())
```

```python
for t in soup.findAll('img'):
    print(t)
```

```python
for t in soup.findAll('img'):
  titles = t.get('title')
  title.append(titles)
```

```python
title
```

```python
filtered_titles = [title for title in title if title and title.strip()]
```

```python
filtered_titles
```

```python
# Filter out unwanted values, including empty strings, newline characters, None, and specific titles
filtered_titles = [title for title in title if title and title.strip() and title not in {'\n\nRead Free Library Books Online\nMillions of books available through Controlled Digital Lending\n\n', '\n\nKeep Track of your Favorite Books\nOrganize your Books using Lists & the Reading Log\n\n', 'Change Website Language'}]
```

```python
filtered_titles
```

Finding all categories

```python
for carousel_section in soup.find_all('div', class_='carousel-section'):
        category_name = carousel_section.find('h2').text

        img_tags = carousel_section.find_all('img')
        for img_tag in img_tags:
            category_names.append(category_name)
```

```python
category_names
```

```python
elements = ['Welcome to Open Library',
 '<a name="Welcome to Open Library"></a>',
 '<a name="Trending Books"></a>',
 '<a name="Classic Books"></a>',
 '<a name="Books We Love"></a>',
 '<a name="Recently Returned"></a>',
 '<a name="Romance"></a>',
 '<a name="Kids"></a>',
 '<a name="Thrillers"></a>',
 '<a name="Textbooks"></a>',
 '<a name="Authors Alliance & MIT Press"></a>',
 '<a name="Browse by Subject"></a>',
 'Welcome to Open Library','Browse by Subject']

for i in elements:
        for j in category_names:
          if i == j:
            category_names.remove(j)
```

```python
len(category_names)
```

#finding all img Urls

```python
for carousel_section in soup.find_all('div', class_='carousel-section'):
        img_tags = carousel_section.find_all('img')
        for img_tag in img_tags:
            img = img_tag.get('src')
            urls_images.append(img)
```

```python
urls_images
```

```python
urls_images = [url for url in urls_images if url.startswith('//covers')]
```

```python
len(urls_images)
```

Finding all Tab_index

```python
for carousel_section in soup.find_all('div', class_='carousel-section'):
        section_tags = carousel_section.find_all('div', class_='cta-button-group')
        for section_tag in section_tags:
          index_tag = section_tag.get('a')
          tab_index.append(index_tag)
```

```python
filtered_tab_index = [item for item in tab_index if item is not None]
filtered_tab_index = [item for sublist in filtered_tab_index for item in sublist]
```

```python
filtered_tab_index
```
